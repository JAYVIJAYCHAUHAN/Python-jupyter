```python
import numpy as np
x= np.arange(0,5,0.1)
y=x**3
y1=x**2
plt.plot(x,y,"b",marker="o",label="x^3")
plt.plot(x,y1,color="red",linewidth=3,label="x^2")
plt.legend()
plt.ylabel("x*3")
plt.xlabel("x")
plt.title("Matplotlib")
plt.grid()
plt.show()
```
