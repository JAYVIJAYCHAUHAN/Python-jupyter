```python
 import matplotlib.pyplot as plt
```

```python
#pie graph
sizes=[2,3,3,2]
plt.title("pie graph")
labels=["a","b","c","d"]
colors=["blue","purple","red","green"]
plt.axis("equal")
explode=[0.1,0,0,0]
plt.pie(sizes,colors=colors,explode=explode,labels=labels,autopct="%.1f%%")
plt.show()
```
