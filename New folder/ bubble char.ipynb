```python
import matplotlib.pyplot as plt
```

```python
year =[2012,2013,2014,2015,2018,2020]
salary=[12,13,14,15,17,20]
popu =(100,120,100, 200 ,500,300)
 
```

```python
plt.scatter(year, salary,s=popu,c="r")
plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year, salary,s=popu,c="r")
plt.title("bubble chart")
plt.show()
```

```python
#aplha decrese color
plt.scatter(year, salary,s=popu,c="r",alpha=0.5)
plt.title("bubble chart")
plt.show()
```

```python
#for different color

import numpy as np
c = np.arange(len(year))
```

```python
plt.scatter(year, salary,s=popu,c=c,alpha=0.6)
plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year, salary,s=popu,c=c,alpha=0.6,marker="^")
plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year, salary,s=popu,c=c,alpha=0.6)
plt.title("bubble chart")
plt.text(year[0]+0.1,salary[0]+0.1,popu[0])
plt.text(year[1]+0.1,salary[1]+0.1,popu[1])
plt.show()
```
